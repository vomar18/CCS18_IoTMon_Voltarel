##  interaction chain discovery algorithm - capability version

this algorithm is based on the table 3 (the correlation between capabilities and channels)

```python
import json   # for read the JSON file
import numpy as np
```

```python
# you need the AP_tr,ac all trigger-action behaviour tuples in applications (cap 4.3.2 - p835)
file = open("../1_application_analysis/2_physical_channel_identification/1_relay_on_capability/all_translated_pair_trigger-action.JSON","r") # read all the capability extrapolated from the apps
translated_trigger_action = json.load(file)
print("tot pair trigger-action:",len(translated_trigger_action))
```

```python
# save all system channels, that in few words are all shared system variables (cap 4.3.1 - p835)
file = open("../1_application_analysis/2_physical_channel_identification/1_relay_on_capability/list_system_cap.JSON","r") # read all the system capabilities extrapolated from the apps
list_system_cap = json.load(file)
list_system_cap
```

```python
# you need the C_ca,ch all relations betweeen capabilities and physical channels (cap 4.3.2 - p835)
file = open("../1_application_analysis/2_physical_channel_identification/1_relay_on_capability/set_of_capability_related_channels.JSON","r")
capability_channel = json.load(file)
capability_channel
```

```python
# algorithm 1: algorithm for interaction chain discovery ***** ONLY PHYSICAL CHANNELS *****
phy_INTAC = []

for i in translated_trigger_action: # for every pair of trigger-action
    for j in translated_trigger_action:
        for channel in capability_channel.keys(): # this implies --> k.ch == m.ch
            #print("physical channel:",physical_channel)
            for k in capability_channel[channel]:     # for every possible pair of capabilities
                for m in capability_channel[channel]: # correlated to the same physical channel
                    action_cap = i[1] # action capability of first trigger-action
                    trigger_cap = j[0]   # trigger capability of second trigger-action
                    if ((action_cap == k) and (trigger_cap == m)):
                        first_app = i  # use only trigger-action
                        second_app = j # use only trigger-action
                        if not [first_app, channel, second_app] in phy_INTAC: # not duplicates 
                            print(f"{action_cap} nel canale {channel} triggera {trigger_cap}")
                            phy_INTAC.append([first_app, channel, second_app])
```

```python
print(len(phy_INTAC)) # for the paper just 570 (cap 5.2 - p839) 
```

```python
# take second element for sorting
def takeSecond(elem):
    return elem[1]

phy_INTAC.sort(key=takeSecond)
phy_INTAC
```

```python
for triple in phy_INTAC:
    first_app = triple[0]
    channel = triple[1]
    second_app = triple[2]
    print(f"tri_1:{first_app[0]}, act_1:{first_app[1]}\t-- {channel} -- tri_2:{second_app[0]}, act:{second_app[1]}")
```

```python
# algorithm 1: algorithm for interaction chain discovery - ALL CHANNELS
sys_INTAC = []

for i in translated_trigger_action: # for every pair of trigger-action
    for j in translated_trigger_action:
        action_cap = i[1]  # action capability of first trigger-action
        trigger_cap = j[0] # trigger capability of second trigger-action
        if action_cap in list_system_cap:
            if action_cap == trigger_cap:
                if not [i, action_cap, j] in sys_INTAC: # not duplicates 
                    print(f"action {action_cap} --channel {action_cap} trigger {trigger_cap}")
                    sys_INTAC.append([i, action_cap, j])
```

```python
print(len(sys_INTAC))
```

```python
for triple in sys_INTAC:
    first_app = triple[0]
    channel = triple[1]
    second_app = triple[2]
    print(f"tri_1:{first_app[0]}, act_1:{first_app[1]}\t-- {channel} -- tri_2:{second_app[0]}, act:{second_app[1]}")
```
